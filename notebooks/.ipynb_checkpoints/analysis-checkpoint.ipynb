```python
def txt2adata(filename='/home/ngr4/scratch60/rpczi/GSE161382_counts_matrix.txt', 
              metadata_filename='/home/ngr4/scratch60/rpczi/GSE161382_metadata.txt',
              save=None):
    '''
    Assumes:
      1. txt file is \t separated
      2. txt file is counts data as genes x cell_id
      3. metadata file is space sepearated, sep=' '
    
    Arguments:
      1. 
    '''
    txtdata = pd.read_csv(filename, sep='\t', index_col=0)
    md = pd.read_csv(metadata_filename, sep=' ')
    adata = sc.AnnData(X=txtdata.T)
    adata.obs = adata.obs.merge(md, left_index=True, right_index=True)
    if save is not None:
        adata.write(save)
    return adata
```

```python
adata.obs['age_in_wk'] = 31
adata.obs.loc[adata.obs['age']=='3yr', 'age_in_wk'] = 3*54
adata.obs.loc[adata.obs['age']=='31yr', 'age_in_wk'] = 31*54
adata.obs.loc[adata.obs['age']=='8yr', 'age_in_wk'] = 8*54
adata.obs.loc[adata.obs['age']=='6mos', 'age_in_wk'] = int(0.5*54)
adata.obs.loc[adata.obs['age']=='7mos', 'age_in_wk'] = int((7/12)*54)
adata.obs.loc[adata.obs['age']=='11mos', 'age_in_wk'] = int((11/12)*54)
adata.obs['age_rangescaled'] = (adata.obs['age_in_wk'] - adata.obs['age_in_wk'].min()) / (adata.obs['age_in_wk'].max() - adata.obs['age_in_wk'].min())
```

```python
# use simple graph 
from scipy import sparse
G = gt.Graph(data=adata.obsp['connectivities']+sparse.diags([1]*adata.shape[0],format='csr'),
                 precomputed='adjacency',
                 use_pygsp=True)
G.knn_max = None
```

```python
import meld
adata.obs['Age (MELD)'] = meld.MELD().fit_transform(G=G, RES=adata.obs['age_rangescaled'])
```

```python
md_var = 'Age (MELD)'
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
im = ax.scatter(data_phate[:, 0], data_phate[:, 1], 
           c=adata.obs[md_var],
           rasterized=True,
                cmap='RdYlBu_r',
           lw=0, s=1, alpha=0.4)
cbar = plt.colorbar(im)
cbar.ax.set_title(md_var)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Dim 1')
ax.set_ylabel('Dim 2')
fig.savefig('/home/ngr4/project/collabs/grants/czi_rp_2103/results/phate_{}.png'.format(md_var), dpi=300, bbox_inches='tight')
```

```python
# REF: https://www.atsjournals.org/doi/full/10.1164/rccm.201912-2338OC
goi = []
ref = ['ACTN3', 'ACTN4', 'CLDN20', 'GSK3B', 'HOXA1', 'HOXB4', 
       'KAT8', 'ITGB5', 'MMP24', 'NCOR2', 'NR3C1', 'ROR2', 
       'RUNX1', 'SERPINC1', 'SOX9', 'WNT2B', 'WNT9A', 'IL6', 'ACE2', 'TMPRSS2']
[goi.append(i) for i in ref if i in adata.var_names.to_list()]
goi
```

```python
import magic
magic_op = magic.MAGIC(t=20).fit(X=adata.X,graph=G)
```

```python
for gene in goi:
    md_var='age'
    fig, ax = plt.subplots(1, 1, figsize=(4, 3))
    sns.scatterplot(x=adata.obs['Age (MELD)'],
                    y=magic_op.transform(adata.X,genes=np.where(adata.var_names==gene)[0]).flatten(),  
                    hue=adata.obs[md_var],
                    rasterized=True,
                    linewidth=0, s=1, alpha=0.4)
    ax.legend(bbox_to_anchor=(1.1, 1))
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel('{} Expression'.format(gene))
    ax.set_xlabel('Age (MELD)')
    fig.savefig('/home/ngr4/project/collabs/grants/czi_rp_2103/results/meldV{}.png'.format(gene), dpi=300, bbox_inches='tight')
```
