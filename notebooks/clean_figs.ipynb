```python
# REF: https://www.atsjournals.org/doi/full/10.1164/rccm.201912-2338OC
goi = []
ref = ['ACTN3', 'ACTN4', 'CLDN20', 'GSK3B', 'HOXA1', 'HOXB4', 
       'KAT8', 'ITGB5', 'MMP24', 'NCOR2', 'NR3C1', 'ROR2', 
       'RUNX1', 'SERPINC1', 'SOX9', 'WNT2B', 'WNT9A', 'IL6', 'ACE2', 'TMPRSS2']
[goi.append(i) for i in ref if i in adata.var_names.to_list()]
goi
```

```python
sc.tl.pca(adata)
sc.external.pp.bbknn(adata, neighbors_within_batch=int(100/adata.obs.batch.unique().shape[0]), n_pcs=50)
```

```python
# use simple graph 
G = gt.Graph(data=adata.obsp['connectivities']+sparse.diags([1]*adata.shape[0],format='csr'),
                 precomputed='adjacency',
                 use_pygsp=True)
G.knn_max = None
```

```python
magic_op = magic.MAGIC(t=20).fit(X=adata.X,graph=G)
for gene in goi:
    md_var='age'
    fig, ax = plt.subplots(1, 1, figsize=(4, 3))
    sns.scatterplot(x=adata.obs['Age (MELD)'],
                    y=magic_op.transform(adata.X,genes=np.where(adata.var_names==gene)[0]).flatten(),  
                    hue=md['age_source'],
                    palette=cmap_age,
                    rasterized=True,
                    linewidth=0, s=1, alpha=0.4)
    ax.legend(bbox_to_anchor=(1.01, 1))
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel('{} Expression'.format(gene))
    ax.set_xlabel('Age (MELD)')
    fig.savefig('/home/ngr4/project/collabs/grants/czi_rp_2103/results/meldVmagic4goi/meldV{}.pdf'.format(gene), dpi=300, bbox_inches='tight')
```
